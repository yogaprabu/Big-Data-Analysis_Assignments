```python
import requests
import lxml
import cssselect
import lxml.html
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from urllib.request import urlopen
import re
```

```python
url = 'https://www.sfu.ca/computing/people/faculty.html'
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
# print(type(soup))
# print(soup)
```

```python
text = soup.get_text()
# print(soup.text)
# print(text)
```

```python
st = soup.find_all('h4')
n=[]
d=[]
for i in range(0,len(st)):
    if(st[i].text.strip()):
        n.append(st[i].text.split(',')[0])
        d.append(st[i].text.split(',')[1:])
# print(n)
s=[]
for i in range(len(d)):
    str1=''.join(str(e) for e in d[i])
    str1=str1.replace('\n','')
    str1=str1.replace('\xa0','')
    if('Area' in str1):
        j=str1.find('Area:')
        str1=str1.replace(str1[j:],'')
    s.append(str1)
# print(d)
```

```python
ao=[]
for i in n:
    j=text.find(i)
    k=text.find('Area: ',j)
    x=text.find('Profile',k)
    if(text[k+6:x].strip()):
        ao.append(text[k:x])
aoi=[]
for i in ao:
    str1=i
    str1=str1.replace('\n','')
    str1=str1.replace('\xa0','')
    if('Area: ' in str1):
        j=str1.find('Area: ')
        str1=str1.replace(str1[j:j+6],'')
    aoi.append(str1)
# print(aoi)
```

```python
data={'Name':n,'Designation':s,'Area Of Interest':aoi}
df=pd.DataFrame(data)
df=df.drop(df.index[range(30,38)])
df=df.drop(df.index[len(df)-1])
df
```
